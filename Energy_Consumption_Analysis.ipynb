```python
columns_to_check = ['Sub Sector', 'Address', 'City', 'Postal Code']

# Drop rows where any value in the specified columns is null
merged_df = merged_df.dropna(subset=columns_to_check)
```

```python
#Drop columns
columns_to_drop = ['Fuel Oil #1 Use (GJ)', 'Fuel Oil #2 Use (GJ)', 'Fuel Oil #4 Use (GJ)', 'Fuel Oil #5 & 6 Use (GJ)']
merged_df = merged_df.drop(columns=columns_to_drop)
```

```python
# Handling missing values
merged_df['Electricity_Quantity'].fillna(merged_df['Electricity_Quantity'].mean(), inplace=True)
merged_df['GHG Emissions KG'].fillna(merged_df['GHG Emissions KG'].mean(), inplace=True)
merged_df['Energy Intensity ekWh_sqft'].fillna(merged_df['Energy Intensity ekWh_sqft'].mean(), inplace=True)
```

```python
# Replace null values in 'NaturalGas_Quantity' with the mode
merged_df['NaturalGas_Quantity'].fillna(merged_df['NaturalGas_Quantity'].mode()[0], inplace=True)
merged_df['FuelOil12_Quantity'].fillna(merged_df['FuelOil12_Quantity'].mode()[0], inplace=True)
merged_df['FuelOil46_Quantity'].fillna(merged_df['FuelOil46_Quantity'].mode()[0], inplace=True)
merged_df['Propane_Quantity'].fillna(merged_df['Propane_Quantity'].mode()[0], inplace=True)
merged_df['Coal_Quantity'].fillna(merged_df['Coal_Quantity'].mode()[0], inplace=True)
merged_df['Wood_Quantity'].fillna(merged_df['Wood_Quantity'].mode()[0], inplace=True)
merged_df['DistrictHeating_Quantity'].fillna(merged_df['DistrictHeating_Quantity'].mode()[0], inplace=True)
merged_df['DistrictCooling_Quantity'].fillna(merged_df['DistrictCooling_Quantity'].mode()[0], inplace=True)
```
